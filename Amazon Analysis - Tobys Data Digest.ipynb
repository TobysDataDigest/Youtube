```python
# General EDA & Data Viz:

# Now I've gone and created some visualiusations to explore further, lets run through them!
```

```python
sns.histplot(df['User Rating'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of User Ratings')
plt.xlabel('User Rating')
plt.ylabel('Frequency')
```

```python
sns.boxplot(x='Genre', y='User Rating', data=df, palette='pastel')
plt.title('User Ratings by Genre')
plt.xlabel('Genre')
plt.ylabel('User Rating')
```

```python
sns.pairplot(df[['User Rating', 'Reviews', 'Price', 'Year']])
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Year', y='User Rating', data=df, palette='coolwarm')
plt.title('User Ratings by Year')
plt.xlabel('Year')
plt.ylabel('User Rating')
```
